# How to deal with missing data

```python
import numpy as np
import pandas as pd
```

```python
d = { 'A':[1,2,np.nan] , 'B':[5, np.nan, np.nan], 'C':[1,2,3] }
```

```python
d
```

```python
df = pd.DataFrame(d)
```

```python
df
```

```python
df.dropna(axis=1)
```

```python
df.dropna()
```

```python
df
```

```python
df.dropna(thresh=2)
```
