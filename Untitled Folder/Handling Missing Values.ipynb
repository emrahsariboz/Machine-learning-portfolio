#### The raw data very likely to include some missing values. We can't feed the data to ML algorithms becasuse it will give us an error. 
#### Thus, we need to find a way to deal with missing values.

#### Dropping the columns if it includes 'NAN' values

```python
import pandas as pd
from sklearn.model_selection import train_test_split

# Read the data
X_full = pd.read_csv('train.csv', index_col='Id')
X_test_full = pd.read_csv('test.csv', index_col='Id')

# Remove rows with missing target, separate target from predictors
X_full.dropna(axis=0, subset=['SalePrice'], inplace=True)
y = X_full.SalePrice
X_full.drop(['SalePrice'], axis=1, inplace=True)

# To keep things simple, we'll use only numerical predictors
X = X_full.select_dtypes(exclude=['object'])
X_test = X_test_full.select_dtypes(exclude=['object'])

# Break off validation set from training data
X_train, X_valid, y_train, y_valid = train_test_split(X, y, train_size=0.8, test_size=0.2,
                                                      random_state=0)
```

```python
## Get the name of columns that includes a missing values
```

```python
col_name= [col for col in X_train.columns if X_train[col].isnull().any()]

col_name
```

## First approach is to drop the NaN values

```python
#Using the drop function, we can drop the columns which includes a missing value
reduced_X_train = X_train.drop(col_name, 1)
reduced_X_valid = X_valid.drop(col_name, 1)
```

## Test the effects of dropping values on RandomForestRegressor

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error

# Function for comparing different approaches
def score_dataset(X_train, X_valid, y_train, y_valid):
    model = RandomForestRegressor(n_estimators=100, random_state=0)
    model.fit(X_train, y_train)
    preds = model.predict(X_valid)
    return mean_absolute_error(y_valid, preds)
```

```python
print("MAE (Drop columns with missing values):")
print(score_dataset(reduced_X_train, reduced_X_valid, y_train, y_valid))
```

## Second approach is to use Simple Impute to impute missing values

```python
from sklearn.impute import SimpleImputer

in_mean = SimpleImputer(strategy='mean')


imputed_X_train = pd.DataFrame(in_mean.fit_transform(X_train))
imputed_X_valid = pd.DataFrame(in_mean.transform(X_valid))

imputed_X_train.columns = X_train.columns
imputed_X_valid.columns = X_valid.columns

imputed_X_train.columns
```

## Test the effects of dropping values on RandomForestRegressor

```python
print("MAE (Imputation):")
print(score_dataset(imputed_X_train, imputed_X_valid, y_train, y_valid))
```

## Third approach is to use Simple Impute (MEDIAN) to impute missing values

```python
# Fill in the lines below: imputation
im_mean = SimpleImputer(strategy='median')

imputed_X_train = pd.DataFrame(im_mean.fit_transform(X_train))
imputed_X_valid = pd.DataFrame(im_mean.transform(X_valid))



# Fill in the lines below: imputation removed column names; put them back
imputed_X_train.columns = X_train.columns
imputed_X_valid.columns = X_valid.columns
```

## Test the effects of dropping values on RandomForestRegressor

```python
print("MAE (Imputation):")
print(score_dataset(imputed_X_train, imputed_X_valid, y_train, y_valid))
```

#### Now, as we can see, this approach is probabilistic. There is not any determined solution to this issue. In our case
#### we found out that replacing the missing values with the median of corresponding column yields better solution
